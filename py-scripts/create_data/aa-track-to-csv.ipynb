```python
import os

import json 
import csv
```

```python
track_files = os.listdir('../audio_analysis/track/')
len(track_files)
```

```python
track_f = open('../audio_analysis/track/{}'.format(track_files[0]))
track_content = json.loads(track_f.read())
```

```python
full_tracks = list()
track_strings = list()
```

```python
for ind, track_file in enumerate(track_files):
    song_id = track_file.split(".")[0]
    
    if ind % 50 == 0:
        print(ind)
    
    track_f = open('../audio_analysis/track/{}'.format(track_file))
    track_content = json.loads(track_f.read())
    track_f.close()
    
    code_map = dict(id=song_id)
    for key in ['codestring', 'code_version', 'echoprintstring', 'echoprint_version', 'synchstring', 'synch_version', 'rhythmstring', 'rhythm_version']:
        if key in track_content.keys():
            code_value = track_content.pop(key)
            code_map[key] = code_value
        else:
            code_map[key] = None
        
    track_content['id'] = song_id
        
    full_tracks.append(track_content)
    track_strings.append(code_map)
```

```python
track_keys = set()

def add_keys_to_set(keys):
    for key in keys:
        track_keys.add(key)
        
added_keys = (add_keys_to_set(track.keys()) for track in full_tracks)
for _ in added_keys: pass

print(track_keys)
```

```python
with open('../audio_analysis/track.csv', 'w') as csvfile:
    fieldnames = list(track_keys)
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(full_tracks)
    
    csvfile.close()
```

```python
with open('../audio_analysis/track-string.csv', 'w') as csvfile:
    fieldnames = [
        'id',
        'codestring', 
        'code_version', 
        'echoprintstring', 
        'echoprint_version', 
        'synchstring', 
        'synch_version', 
        'rhythmstring', 
        'rhythm_version'
    ]
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(track_strings)
    
    csvfile.close()
```
