```python
import csv
import requests
import json
```

```python
spotify_ids = set()
with open('../../data/ranks/song-ranked-analytics.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['spotify_id']:
            spotify_ids.add(''.join(['spotify:track:', row['spotify_id']]))
            
spotify_ids = list(spotify_ids)
```
