```python
import os

import json 
import csv

from uuid import uuid4
from copy import deepcopy
```

```python
segments_f = open('../audio_analysis/segments/{}'.format(segments_files[0]))
segments_contents = json.loads(segments_f.read())
```

>This tells me that the length of both the pitches and timbre field are a set length, and likely the same for every song analysis. I am going to extract each one of these fields from the segments field, and give each segment a unique identifier to map them to.

```python
pitches_lens = set()
timbre_lens = set()

for segment in segments_contents:
    pitches_lens.add(len(segment['pitches']))
    timbre_lens.add(len(segment['timbre']))
    
print(pitches_lens)
print(timbre_lens)
```

```python
wrong_segments_type = list()
full_segments = list()

wrong_pitches_type = list()
full_pitches = list()

wrong_timbres_type = list()
full_timbres = list()
```

```python
for ind, segments_file in enumerate(segments_files):
    song_id = segments_file.split(".")[0]
    
    if ind % 50 == 0:
        print(ind)
    
    segment_f = open('../audio_analysis/segments/{}'.format(segments_file))
    segment_content = json.loads(segment_f.read())
    segment_f.close()
    
    if isinstance(segment_content, list):
        for segment in segment_content:
            segment_id = str(uuid4())
            
            segment['id'] = segment_id
            segment['song_id'] = song_id
            
            segment_copy = deepcopy(segment)
            
            if 'pitches' in segment.keys():
                segment.pop('pitches')
                
            if 'timbre' in segment.keys():
                segment.pop('timbre')
                        
            segment_pitches = segment_copy.get('pitches')
            segment_timbre = segment_copy.get('timbre')
            
            if segment_pitches and isinstance(segment_pitches, list):
                mapped_pitches = {ind: pitch for ind, pitch in enumerate(segment_pitches)}
                mapped_pitches['segment_id'] = segment_id
                mapped_pitches['song_id'] = song_id
                
                full_pitches.extend([mapped_pitches])
            else:
                wrong_pitches_type.append(song_id)
                
            if segment_timbre and isinstance(segment_timbre, list):
                mapped_timbre = {ind: timbre for ind, timbre in enumerate(segment_timbre)}
                mapped_timbre['segment_id'] = segment_id
                mapped_timbre['song_id'] = song_id
                
                full_timbres.extend([mapped_timbre])
            
            else: 
                wrong_timbres_type.append(song_id)

    else:
        wrong_segments_type.append(segments_file)
        
    full_segments.extend(segment_content)
```

```python
segment_keys = set()

for segment in full_segments:
    for key in segment.keys():
        segment_keys.add(key)
```

```python
segment_keys = list(segment_keys)

for segment in full_segments:
    for key in segment_keys:
        if key not in segment.keys():
            segment[key] = None
```

```python
with open('../audio_analysis/segments.csv', 'w') as csvfile:
    fieldnames = [
        'id', 
        'song_id',
        'confidence', 
        'start', 
        'duration',         
        'loudness_max', 
        'loudness_max_time', 
        'loudness_start',
        'loudness_end'
    ]
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(full_segments)
    
    csvfile.close()
```
