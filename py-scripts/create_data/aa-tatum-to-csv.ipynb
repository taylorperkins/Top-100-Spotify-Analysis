```python
import os

import json 
import csv
```

```python
tatums_files = os.listdir('../audio_analysis/tatums/')
len(tatums_files)
```

```python
tatums_f = open('../audio_analysis/tatums/{}'.format(tatums_files[0]))
tatums_content = json.loads(tatums_f.read())
```

```python
tatums_content
```

```python
wrong_tatums_type = list()
full_tatums = list()
```

```python
for ind, tatums_file in enumerate(tatums_files):
    song_id = tatums_file.split(".")[0]
    
    if ind % 50 == 0:
        print(ind)
    
    tatums_f = open('../audio_analysis/tatums/{}'.format(tatums_file))
    tatums_content = json.loads(tatums_f.read())
    tatums_f.close()
    
    if isinstance(tatums_content, list):
        for tatum in tatums_content:
            tatum['id'] = song_id
    else:
        wrong_tatums_type.append(tatums_file)
        
    full_tatums.extend(tatums_content)
```

```python
wrong_tatums_type
```

```python
full_tatums[0]
```

```python
with open('../audio_analysis/tatums.csv', 'w') as csvfile:
    fieldnames = [
        'id', 
        'confidence', 
        'start',
        'duration' 
    ]
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(full_tatums)
    
    csvfile.close()
```
