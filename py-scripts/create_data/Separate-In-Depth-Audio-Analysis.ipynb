```python
import os
import json

import pandas as pd
```

```python
audio_files = os.listdir('../audio_analysis/')
```

```python
f = open('../audio_analysis/' + audio_files[0], 'r')
song_analysis = json.loads(f.read())
f.close()
```

```python
song_analysis.keys()
```

```python
incorrect_files = list()
```

```python
def check_for_keys(song_analysis, file_name):
    for key in ['meta', 'track', 'bars', 'beats', 'tatums', 'sections', 'segments']:
        if not key in song_analysis.keys():
            print('Key not found')
            del song_analysis
            incorrect_files.append(file_name)
            return
    del song_analysis
    return
```

```python
for ind, file_name in enumerate(audio_files):
    if ind % 50 == 0:
        print(ind)
        
    f = open('../audio_analysis/' + file_name, 'r')
    song_analysis = json.loads(f.read())
    f.close()
    
    check_for_keys(song_analysis, file_name)
```

```python
for ind, file_name in enumerate(audio_files):
    song_id = file_name.split("-")[0]
    
    if ind % 50 == 0:
        print(ind)
        
    f = open('../audio_analysis/full-analysis/' + file_name, 'r')
    song_analysis = json.loads(f.read())
    f.close()
    
    for key in ['meta', 'track', 'bars', 'beats', 'tatums', 'sections', 'segments']:
        song_data = song_analysis.get(key)
        
        data_file = open('../audio_analysis/{}/{}.json'.format(key, song_id), 'w')
        data_file.write(json.dumps(song_data))
        data_file.close()
```
