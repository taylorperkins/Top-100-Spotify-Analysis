```python
def from_json_to_csv(path, data_type):
    data_set = list()
    dir_path = path + '/{}/'.format(data_type)
    
    for file in os.listdir(dir_path):
        if file.endswith('.json'):
            data_file = os.path.join(dir_path, file)
            print(data_file)
            
            rank = None
            if '-' in file:
                rank = file.split('-')[0]
            
            with open(data_file, 'r') as f:
                try: 
                    file_contents = json.loads(f.read())
                except json.JSONDecodeError as e:
                    print('\n\n{}\n\n'.format(e))
                    continue
                
                for row in file_contents:
                    row['rank'] = rank
                                
                data_set.extend(file_contents)
                print("Extended the set!! {}".format(file_contents[0]))
                
            f.close()
            
    print(data_set[1:5])
    
    headers = data_set[0].keys()
    csvfile = _current_dir + '/{}.csv'.format(data_type)
    with open(csvfile, 'w') as cf:        
        writer = csv.DictWriter(cf, fieldnames=headers)

        writer.writeheader()
        writer.writerows(data_set)
```
